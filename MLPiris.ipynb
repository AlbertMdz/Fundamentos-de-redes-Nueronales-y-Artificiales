```
class Capas_redneuronal():

    def __init__(self, n_conn, n_neur, act_f):
        self.act_f = act_f

        self.b = np.random.rand(1, n_neur) * 2 - 1
        self.W = np.random.rand(n_conn, n_neur) * 2 - 1
```

```
def creaciondered(modelo, funciones):
    FF = []
    for l, layer in enumerate(modelo[:-1]):
        FF.append(Capas_redneuronal(modelo[l], modelo[l + 1], funciones))

    return FF
```
